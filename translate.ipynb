```python
import pandas as pd
import numpy as np
```

```python
import json
```

```python
f = open('sp_en_dict.json', 'r', encoding='ascii')
data = json.load(f)
data = { data[k]:k for k in data}
data
```

```python
import re
new_dic = {}
```

```python
import unidecode
```

```python
for key in data:
    val = data[key]
    new_key = re.sub('{.*?}', '', key)
    possible_words = new_key.split(', ')
    possible_words = [w.strip() for w in possible_words]
    for possible_word in possible_words:
        new_dic[unidecode.unidecode(possible_word)] =  val
```

```python
new_dic.pop('', None)
new_dic
```

```python
df = pd.read_csv('corpus.csv')
df
```

```python
import nltk
nltk.download('words')
from nltk.corpus import words
```

```python
en_words = set()
for i in words.words():
    en_words.add(i)
len(en_words)
```

```python
def trans(line):
    arr = line.split(' ')
    new_arr = []
    for word in arr:
        if word in new_dic.keys():
            new_arr.append(new_dic[word])
        else:
            new_arr.append(word)
    return ' '.join(new_arr)
```

```python
new_text = []
for index, row in df.iterrows():
    try:
      new_text.append(trans(row.tweet))
    except:
      new_text.append('')
    if index%250 == 0:
      print(index)
# df['text'] = df['text'].apply(lambda line: trans(line))
```
